```python
def chatbot(question) :
    user_question= question.lower() 
    if "hello" in user_question or "hi" in user_question:
        return "hello, nice to talk to you"
    elif "how are you" in user_question:
        return"im good, how are you"
    elif "im good" in user_question or "im alright" in user_question:
        return "good to hear that"
    elif "joke" in user_question:
        return "why did the computer go to the doctor? becouse it got a virus"
    else:
        return "i didnt understand that sorry"
print("chatbot is ready, type somthing below") 
while True:
    


    my_question=input("you: ") 
    if my_question.lower() in ["bye","quit","exit"]:
        print("Goodbye")
        break
    
    print("bot:",chatbot(my_question))

```
